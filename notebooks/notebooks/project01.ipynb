```python
import pandas as pd

df = pd.read_csv('01_fund_master.csv')

print("Shape:")
print(df.shape)

print("\nData Types:")
print(df.dtypes)

print("\nFirst 5 Rows:")
print(df.head())
print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicates:")
print(df.duplicated().sum())

print("\nSummary Statistics:")
print(df.describe())
```
